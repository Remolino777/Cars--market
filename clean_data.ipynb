```python
# Creo un df con las columnas de variables categoricas para hacer una exploracion de sus valores unicos.

df_c = df.select_dtypes(include=['object']).copy()
df_c.info()
```

```python
nombres_columnas_df_c = df_c.columns.to_list() # copio los nombres de las columnas a una lista

for columna in nombres_columnas_df_c:  # Creo el bucle para ver la imformacion de cada columna
    cc_u = df_c[columna].value_counts() 
    print(cc_u)
```

```python
# reviso los valores de la columna de 'CarName' ya que es la mas extensa y puede haver un error 

conteo_car_names = df_c['CarName'].value_counts(dropna=False).sort_index()
for index, value in conteo_car_names.items():
    print(index, value)
```

```python
df1 = df.copy() # Hago una copia del df original
```

#### Por ser muchas columnas y para tener mejor facilidad de observar los outlayers en la grafica dividire los valores discretos en dos grupos.

```python
df_d = df1.select_dtypes(exclude=['object']) # Creo una copia de los df que excluyan las columnas de tipo 'object'
```

```python
len(df_d.columns) # verifico que el total de columnas sea par 
```

```python
# Dividir las columnas del DataFrame original en dos grupos
mitad_columnas = len(df_d.columns) // 2
grupo1_columnas = df_d.columns[:mitad_columnas]
grupo2_columnas = df_d.columns[mitad_columnas:]
```

```python
#Creo los nuevos df
df_d1 = df_d[grupo1_columnas].copy()
df_d2 = df_d[grupo2_columnas].copy()
```

```python
df_d1
```

```python
df_d2
```

#### Grafico los df

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'width_ratios':[2,1]})


df_d1[df_d1.columns[df_d1.columns != 'curbweight']].boxplot(ax=axs[0])
df_d1['curbweight'].plot(kind='box', ax=axs[1])


#titulos
axs[0].set_title('Datos df_d1')

axs[1].set_title('Boxplot de enginesize, curbweight')



plt.tight_layout()
plt.show()
```

```python
# Verifico los valores del curbweight y enginesize de lso siguientes carros 
nuevo_df = df.loc[(df_d1['enginesize'] >= 200) & (df_d1['enginesize'] <= 400), ['car_ID','curbweight', 'enginesize']]
nuevo_df
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'width_ratios':[2,1]})


df_d2[df_d2.columns[df_d2.columns != 'price']].boxplot(ax=axs[0])
df_d2['price'].plot(kind='box', ax=axs[1])


#titulos
axs[0].set_title('Datos df_d2')

axs[1].set_title('price, curbweight')



plt.tight_layout()
plt.show()
```

```python
# Reviso la columna 'precio' para ver cual es el carro que se sale de la media
df_p = df.loc[df['price'] > 40000]
df_p
```

#### Elimino los outlayers !!!!!!! Pero del dataframe donde estan todas las columnas "df1" !!!!!

```python

indices_mayor_45 = df1[df1['price'] > 40000].index
peak = df1[df['peakrpm'] > 6000].index
df1 = df1.drop(indices_mayor_45)
df1 = df1.drop(peak)
```
