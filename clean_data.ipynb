```python
# Elimino las columnas de CarNamey car_id ya que no aporta nada al modelo y hago una copia del original
df1 = df.drop(columns=['CarName', 'car_ID']).copy() # Hago una copia del df original
df1
```

#### Por ser muchas columnas y para tener mejor facilidad de observar los outlayers en la grafica dividire los valores discretos en dos grupos.

```python
df_d = df1.select_dtypes(exclude=['object']) # Creo una copia de los df que excluyan las columnas de tipo 'object'
```

```python
len(df_d.columns) # verifico que el total de columnas sea par 
```

```python
# Dividir las columnas del DataFrame original en dos grupos
mitad_columnas = len(df_d.columns) // 2
grupo1_columnas = df_d.columns[:mitad_columnas]
grupo2_columnas = df_d.columns[mitad_columnas:]
```

```python
#Creo los nuevos df
df_d1 = df_d[grupo1_columnas].copy()
df_d2 = df_d[grupo2_columnas].copy()
```

#### Grafico los df

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'width_ratios':[2,1]})


df_d1[df_d1.columns[df_d1.columns != 'curbweight']].boxplot(ax=axs[0])
df_d1['curbweight'].plot(kind='box', ax=axs[1])


#titulos
axs[0].set_title('Datos df_d1')

axs[1].set_title('Boxplot de enginesize, curbweight')



plt.tight_layout()
plt.show()
```

#### Elimino los outlayers !!!!!!! Pero del dataframe donde estan todas las columnas "df1" !!!!!

```python

indices_mayor_45 = df1[df1['price'] > 40000].index
peak = df1[df['peakrpm'] > 6000].index
df1 = df1.drop(indices_mayor_45)
df1 = df1.drop(peak)
```

```python
# Verifico que los outlayer ya no estan

fig, axs = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'width_ratios':[2,1]})


df1[df1.columns[df1.columns != 'price']].boxplot(ax=axs[0])
df1['price'].plot(kind='box', ax=axs[1])


#titulos
axs[0].set_title('Datos df1')

axs[1].set_title('price, curbweight')



plt.tight_layout()
plt.show()
```

#### Utilizo la tecnica de transformacion logaritmica Estas nuevas características tienen la propiedad de que reducen la escala de los valores extremos y pueden hacer que los datos sean más adecuados ...

```python
import numpy as np
```

```python
### Aplicamos trasnformacion logaritmica para los valores 'price', 'enginesize', 'curbweight'

df1['price_log'] = np.log(df1['price'])
df1['enginesize_log'] = np.log(df1['enginesize'])
df1['curbweight_log'] = np.log(df1['curbweight'])

df1_log = df1[['curbweight_log', 'enginesize_log','price_log']]
df1_log
```

```python
df1_log.boxplot()

plt.title('Datos df1 (Transformado)')


plt.show()
```
